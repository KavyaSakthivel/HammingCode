```python
def generate_hamming_code(data_word):
    # number of parity bits required
    r = 0
    while 2**r < len(data_word) + r + 1:
        r += 1
    
    #  Hamming code with parity bits set to 0
    hamming_code = [0] * (len(data_word) + r)
    j = 0
    
    for i in range(len(hamming_code)):
        if i+1 not in [2**x for x in range(r)]:
            hamming_code[i] = int(data_word[j])
            j += 1
    
    for i in range(r):
        parity_index = 2**i - 1
        count = 0
        for j in range(len(hamming_code)):
            if (j+1) & (1 << i):
                count += hamming_code[j]
        if count % 2 == 0:
            hamming_code[parity_index] = 0
        else:
            hamming_code[parity_index] = 1
            
    return hamming_code

def decode_hamming_code(received_code):
    r = 0
    while 2**r < len(received_code):
        r += 1
    
    # Initialize error position and syndrome
    error_pos = -1  # Initialize with -1 to indicate no error found
    syndrome = 0
    
    # syndrome
    for i in range(r):
        parity_index = 2**i - 1
        count = 0
        for j in range(len(received_code)):
            if (j+1) & (1 << i):
                count += received_code[j]
        if count % 2 != received_code[parity_index]:
            syndrome += 2**i
    
    # error check
    if syndrome != 0:
        error_pos = syndrome - 1
    
    # Correct error if any
    if error_pos != -1:
        received_code[error_pos] ^= 1
    
    # Extract and return the data bits
    data_word = ""
    j = 0
    for i in range(len(received_code)):
        if i+1 not in [2**x for x in range(r)]:
            data_word += str(received_code[i])
    
    return data_word
#input
data_word = input("Enter the data word (7 bits): ")
print("User input (data word):", data_word)

# Hamming code
hamming_code = generate_hamming_code(data_word)
print("Hamming code:", hamming_code)

# Decoded
decoded_data_word = decode_hamming_code(hamming_code)
print("Decoded data word:", decoded_data_word)
```
